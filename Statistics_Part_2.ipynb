Q1)  Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

```python
import numpy as np
from scipy import stats

# Sample data
sample_mean = 25.5
sample_std = 3.2
sample_size = 36

# Known population mean
population_mean = 24.8
population_std = 3.5

# Calculate the standard error
standard_error = population_std / np.sqrt(sample_size)

# Calculate the Z-score
z_score = (sample_mean - population_mean) / standard_error

# Calculate the P-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print the results
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

 Q5) Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_errors(mu0, mu1, sigma, alpha, sample_size):
    # Calculate the critical value for Type 1 error
    critical_value = norm.ppf(1 - alpha, loc=mu0, scale=sigma / np.sqrt(sample_size))

    # Calculate Type 1 error (alpha)
    type1_error = alpha

    # Calculate Type 2 error (beta)
    type2_error = norm.cdf(critical_value, loc=mu1, scale=sigma / np.sqrt(sample_size))

    return type1_error, type2_error, critical_value

def visualize_errors(mu0, mu1, sigma, alpha, sample_size, critical_value):
    x = np.linspace(mu0 - 3 * sigma / np.sqrt(sample_size), mu1 + 3 * sigma / np.sqrt(sample_size), 100)

    # Plot null hypothesis distribution
    plt.plot(x, norm.pdf(x, loc=mu0, scale=sigma / np.sqrt(sample_size)), label='Null Hypothesis')

    # Plot alternative hypothesis distribution
    plt.plot(x, norm.pdf(x, loc=mu1, scale=sigma / np.sqrt(sample_size)), label='Alternative Hypothesis')

    # Shade Type 1 error region
    plt.fill_between(x, norm.pdf(x, loc=mu0, scale=sigma / np.sqrt(sample_size)), where=(x > critical_value), alpha=0.3, label='Type 1 Error')

    # Shade Type 2 error region
    plt.fill_between(x, norm.pdf(x, loc=mu1, scale=sigma / np.sqrt(sample_size)), where=(x < critical_value), alpha=0.3, label='Type 2 Error')

    # Plot critical value
    plt.axvline(x=critical_value, color='red', linestyle='--', label='Critical Value')

    plt.legend()
    plt.show()

# Example usage
mu0 = 0  # Null hypothesis mean
mu1 = 1  # Alternative hypothesis mean
sigma = 1  # Standard deviation
alpha = 0.05  # Significance level
sample_size = 100  # Sample size

type1_error, type2_error, critical_value = calculate_errors(mu0, mu1, sigma, alpha, sample_size)
print(f"Type 1 Error: {type1_error:.4f}")
print(f"Type 2 Error: {type2_error:.4f}")
print(f"Critical Value: {critical_value:.4f}")

visualize_errors(mu0, mu1, sigma, alpha, sample_size, critical_value)
```

Q9) Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
import numpy as np
from scipy import stats

def calculate_confidence_interval(sample_data, confidence_level=0.95):
    # Calculate sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)  # Use sample standard deviation

    # Calculate standard error
    standard_error = sample_std / np.sqrt(len(sample_data))

    # Calculate critical value from t-distribution
    degrees_of_freedom = len(sample_data) - 1
    critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    # Calculate margin of error
    margin_of_error = critical_value * standard_error

    # Calculate confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage
sample_data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 26, 20])
confidence_level = 0.95
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"{confidence_level*100}% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
```

Q10) Write a Python program to calculate the margin of error for a given confidence level using sample data

```python
import numpy as np
from scipy import stats

def calculate_margin_of_error(sample_data, confidence_level=0.95):
    # Calculate sample standard deviation
    sample_std = np.std(sample_data, ddof=1)

    # Calculate standard error
    standard_error = sample_std / np.sqrt(len(sample_data))

    # Calculate critical value from t-distribution
    degrees_of_freedom = len(sample_data) - 1
    critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    # Calculate margin of error
    margin_of_error = critical_value * standard_error

    return margin_of_error

# Example usage
sample_data = np.array([23, 21, 19, 24, 20, 22, 18, 25, 26, 20])
confidence_level = 0.95
margin_of_error = calculate_margin_of_error(sample_data, confidence_level)

print(f"Sample Mean: {np.mean(sample_data):.4f}")
print(f"Margin of Error at {confidence_level*100}% Confidence Level: {margin_of_error:.4f}")
print(f"Confidence Interval: ({np.mean(sample_data) - margin_of_error:.4f}, {np.mean(sample_data) + margin_of_error:.4f})")
```

Q11)  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
import numpy as np

def bayes_theorem(prior_A, prob_B_given_A, prob_B):
    # Calculate posterior probability using Bayes' Theorem
    posterior_A_given_B = (prob_B_given_A * prior_A) / prob_B
    return posterior_A_given_B

# Example usage
# Let's say we want to determine the probability of a person having a disease (A) given that they tested positive (B)
prior_A = 0.01  # Prior probability of having the disease (1%)
prob_B_given_A = 0.99  # Probability of testing positive given that the person has the disease (99% accurate test)
prob_B_given_not_A = 0.05  # Probability of testing positive given that the person does not have the disease (5% false positive rate)
prob_not_A = 1 - prior_A  # Probability of not having the disease

# Calculate probability of testing positive (B)
prob_B = (prior_A * prob_B_given_A) + (prob_not_A * prob_B_given_not_A)

# Calculate posterior probability using Bayes' Theorem
posterior_A_given_B = bayes_theorem(prior_A, prob_B_given_A, prob_B)

print(f"Posterior Probability of having the disease given a positive test result: {posterior_A_given_B:.4f}")
```

Q13)  Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

```python
import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    # Calculate row and column totals
    row_totals = observed_data.sum(axis=1)
    col_totals = observed_data.sum(axis=0)
    grand_total = observed_data.sum().sum()

    # Calculate expected frequencies
    expected_frequencies = np.outer(row_totals, col_totals) / grand_total

    return expected_frequencies

# Create a contingency table
observed_data = np.array([[20, 15], [10, 25]])
df = pd.DataFrame(observed_data, index=['Group A', 'Group B'], columns=['Category 1', 'Category 2'])

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(df)

# Print observed and expected frequencies
print("Observed Frequencies:")
print(df)
print("\nExpected Frequencies:")
print(expected_frequencies)
```

Q16)  Implement an F-test using Python to compare the variances of two random samples

```python
import numpy as np
from scipy import stats

# Generate two random samples
np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=1.5, size=100)

# Perform F-test
f_statistic, p_value = stats.f_oneway(sample1, sample2)

# However, for comparing variances directly, we can calculate the F-statistic manually
var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)
f_statistic_manual = var1 / var2 if var1 > var2 else var2 / var1
p_value_manual = 2 * (1 - stats.f.cdf(f_statistic_manual, len(sample1) - 1, len(sample2) - 1))

print(f"F-statistic (manual calculation): {f_statistic_manual:.4f}")
print(f"P-value (manual calculation): {p_value_manual:.4f}")

# Interpret the results
alpha = 0.05
if p_value_manual < alpha:
    print("Reject the null hypothesis. The variances are likely to be different.")
else:
    print("Fail to reject the null hypothesis. The variances are likely to be equal.")
```

Q19)  Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

def check_anova_assumptions(data):
    # Check normality assumption
    print("Normality Assumption:")
    for i, group in enumerate(data):
        stat, p = stats.shapiro(group)
        print(f"Group {i+1}: p-value = {p:.4f}")
        if p < 0.05:
            print(f"Group {i+1} does not appear to be normally distributed.")
        else:
            print(f"Group {i+1} appears to be normally distributed.")

    # Check equal variance assumption
    print("\nEqual Variance Assumption:")
    stat, p = stats.levene(*data)
    print(f"p-value = {p:.4f}")
    if p < 0.05:
        print("The groups do not appear to have equal variances.")
    else:
        print("The groups appear to have equal variances.")

    # Check independence assumption
    # Note: Independence is usually ensured by the study design.
    # Here, we can't directly test for independence without more information.
    print("\nIndependence Assumption:")
    print("Independence is assumed to be ensured by the study design.")

    # Plot histograms for normality check
    plt.figure(figsize=(10, 6))
    for i, group in enumerate(data):
        plt.subplot(1, len(data), i+1)
        plt.hist(group, bins=10, alpha=0.5, label=f"Group {i+1}")
        plt.title(f"Group {i+1}")
        plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(0)
group1 = np.random.normal(loc=5, scale=1, size=100)
group2 = np.random.normal(loc=6, scale=1, size=100)
group3 = np.random.normal(loc=7, scale=1, size=100)
check_anova_assumptions([group1, group2, group3])
```

Q20)  Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

```python
import numpy as np
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Sample data
np.random.seed(0)
factor1 = np.repeat(['A', 'B'], 20)
factor2 = np.tile(['X', 'Y'], 20)
values = np.random.normal(loc=5, scale=1, size=40)

# Add interaction effect
for i in range(len(values)):
    if factor1[i] == 'A' and factor2[i] == 'X':
        values[i] += 2
    elif factor1[i] == 'B' and factor2[i] == 'Y':
        values[i] += 1

# Create a DataFrame
df = pd.DataFrame({
    'Factor1': factor1,
    'Factor2': factor2,
    'Value': values
})

# Perform two-way ANOVA test
model = ols('Value ~ C(Factor1) + C(Factor2) + C(Factor1):C(Factor2)', data=df).fit()
anova_table = anova_lm(model, typ=2)

print(anova_table)

# Visualize the results
plt.figure(figsize=(8, 6))
sns.boxplot(x='Factor1', y='Value', hue='Factor2', data=df)
plt.title("Boxplot of Values")
plt.show()

# Interaction plot
plt.figure(figsize=(8, 6))
sns.pointplot(x='Factor1', y='Value', hue='Factor2', data=df)
plt.title("Interaction Plot")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Define the degrees of freedom
df1 = 5
df2 = 10

# Generate x values
x = np.linspace(0, 5, 100)

# Calculate the probability density function (pdf)
y = stats.f.pdf(x, df1, df2)

# Plot the F-distribution
plt.plot(x, y)
plt.title(f"F-Distribution (df1 = {df1}, df2 = {df2})")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.show()

# Plot F-distributions with different degrees of freedom
df1_values = [1, 5, 10]
df2_values = [10, 10, 10]

plt.figure(figsize=(10, 6))
for df1 in df1_values:
    y = stats.f.pdf(x, df1, df2_values[df1_values.index(df1)])
    plt.plot(x, y, label=f"df1 = {df1}, df2 = {df2_values[df1_values.index(df1)]}")

plt.title("F-Distributions with Different Degrees of Freedom")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()

```

Q23)  Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results

```python
import numpy as np
from scipy import stats

# Sample data
np.random.seed(0)
data = np.random.normal(loc=5, scale=1, size=100)

# Calculate sample variance
sample_variance = np.var(data, ddof=1)

# Null hypothesis: population variance is equal to 1
null_hypothesis_variance = 1

# Calculate Chi-square statistic
n = len(data)
chi_square_statistic = (n - 1) * sample_variance / null_hypothesis_variance

# Calculate p-value
p_value = 2 * min(stats.chi2.cdf(chi_square_statistic, n - 1), 1 - stats.chi2.cdf(chi_square_statistic, n - 1))

print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The population variance is likely to be different from the hypothesized value.")
else:
    print("Fail to reject the null hypothesis. The population variance is likely to be equal to the hypothesized value.")
```

Q24)  Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python
import numpy as np
from scipy import stats

# Sample data
successes1 = 120
sample_size1 = 200
successes2 = 150
sample_size2 = 250

# Calculate proportions
proportion1 = successes1 / sample_size1
proportion2 = successes2 / sample_size2

# Calculate pooled proportion
pooled_proportion = (successes1 + successes2) / (sample_size1 + sample_size2)

# Calculate standard error
standard_error = np.sqrt(pooled_proportion * (1 - pooled_proportion) * (1 / sample_size1 + 1 / sample_size2))

# Calculate Z-statistic
z_statistic = (proportion1 - proportion2) / standard_error

# Calculate p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print(f"Proportion 1: {proportion1:.4f}")
print(f"Proportion 2: {proportion2:.4f}")
print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The proportions are likely to be different.")
else:
    print("Fail to reject the null hypothesis. The proportions are likely to be equal.")

```

Q26)  Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results


```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Sample data
np.random.seed(0)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=0, scale=1.5, size=100)

# Calculate variances
variance1 = np.var(data1, ddof=1)
variance2 = np.var(data2, ddof=1)

# Perform F-test
f_statistic = variance1 / variance2 if variance1 > variance2 else variance2 / variance1
df1 = len(data1) - 1 if variance1 > variance2 else len(data2) - 1
df2 = len(data2) - 1 if variance1 > variance2 else len(data1) - 1
p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

print(f"Variance 1: {variance1:.4f}")
print(f"Variance 2: {variance2:.4f}")
print(f"F-statistic: {f_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The variances are likely to be different.")
else:
    print("Fail to reject the null hypothesis. The variances are likely to be equal.")

# Visualize the results
plt.hist(data1, bins=30, alpha=0.5, label='Data 1')
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.legend()
plt.show()

# Visualize the F-distribution
x = np.linspace(0, 5, 100)
y = stats.f.pdf(x, df1, df2)
plt.plot(x, y)
plt.axvline(x=f_statistic, color='r', linestyle='dashed', label='F-statistic')
plt.legend()
plt.show()
```

Q27)  Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Simulate data from a normal distribution
np.random.seed(0)
data = np.random.normal(loc=5, scale=2, size=1000)

# Define the bins for the histogram
bins = np.linspace(0, 10, 11)

# Calculate the observed frequencies
observed_frequencies, _ = np.histogram(data, bins=bins)

# Calculate the expected frequencies under a normal distribution
expected_frequencies = stats.norm.cdf(bins[1:], loc=5, scale=2) - stats.norm.cdf(bins[:-1], loc=5, scale=2)
expected_frequencies *= len(data)

# Perform Chi-square test
chi_square_statistic = np.sum((observed_frequencies - expected_frequencies) ** 2 / expected_frequencies)
df = len(observed_frequencies) - 1 - 2  # 2 parameters (mean and std) are estimated
p_value = 1 - stats.chi2.cdf(chi_square_statistic, df)

print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The data does not fit the normal distribution.")
else:
    print("Fail to reject the null hypothesis. The data fits the normal distribution.")

# Visualize the results
plt.hist(data, bins=bins, alpha=0.5, label='Observed Frequencies')
plt.plot((bins[1:] + bins[:-1]) / 2, expected_frequencies, 'ro-', label='Expected Frequencies')
plt.legend()
plt.show()
```
